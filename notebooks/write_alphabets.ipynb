```python
import pandas as pd
```

```python
# Downloaded the table from 
# https://loc.gov/marc/specifications/specchareacc/JapaneseHiraganaKatakana.html
df = pd.read_csv('../alphabets/alphabet.csv')
```

```python
df.head()
```

```python
# remove the last 2 vocalization symbols from table
df = df[:-2]
```

```python
df_hiragana = df[df['NAME'].str.contains("Hiragana")]
print(df_hiragana[~df_hiragana['NAME'].str.contains("small")])
```

```python
df_katakana = df[df['NAME'].str.contains("Katakana")]
print(df_katakana[~df_katakana['NAME'].str.contains("small")])
```

```python
hiragana_alphabet = pd.DataFrame()
hiragana_alphabet['CHAR'] = df_hiragana[~df_hiragana['NAME'].str.contains("small")]['CHAR']
hiragana_alphabet['NAME'] = df_hiragana['NAME'].str.rsplit(' ').str[-1]
hiragana_alphabet['NAME'] = hiragana_alphabet['NAME'].str.lower()
hiragana_alphabet
```

```python
katakana_alphabet = pd.DataFrame()
katakana_alphabet['CHAR'] = df_katakana[~df_katakana['NAME'].str.contains("small")]['CHAR']
katakana_alphabet['NAME'] = df_katakana['NAME'].str.rsplit(' ').str[-1]
katakana_alphabet['NAME'] = katakana_alphabet['NAME'].str.lower()
katakana_alphabet
```

```python
hiragana_alphabet.to_csv('../alphabets/hiragana.csv', index=False)
katakana_alphabet.to_csv('../alphabets/katakana.csv', index=False)
```
